```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(country, population, width=0.5, color= ['#5c58b6','#b457ce','#5994ce','#3a4e93'])
#configuraciones eje x
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'),
           rotation = 45)
plt.title('Habitantes por pais')
plt.xlabel('Paises')
plt.ylabel('Habitantes')
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
