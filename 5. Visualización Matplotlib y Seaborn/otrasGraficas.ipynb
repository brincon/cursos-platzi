```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
# bin, numero de barras
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
# bin, numero de barras
plt.hist(data, bins=10, histtype='step')
plt.show()
```

## Boxplot

```python
# 200 es un outlier, los datos están de 0-50
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
# 200 es un outlier, los datos están de 0-50
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30* np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
# alpha nos muestra transparencia y eso deja ver que puntos 
# estàn sobre otros
plt.scatter(x,y, s=area, c=colors, marker='p', alpha=0.5)
plt.show()
```
