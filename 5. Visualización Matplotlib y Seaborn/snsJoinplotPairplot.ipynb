```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

Une dos gráficas en una sola. Tenemos un diagrama de dispersión y una distribucón en los ejes x y y.

```python
#kind= hist, kde
# marginal_ticks = tabla de la escala de los histogramas
sns.jointplot(data=tips, x='total_bill', y='tip', hue="sex", kind='hist',
              marginal_ticks=True, marginal_kws=dict(bins=25,fill=False,
                                                     multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True,
             kind='scatter')
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()
```
