```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
# metodo no orientado a objetos
# por defecto lo ubica donde lo crea m√°s conveniente pero se le puede decir
# con loc la ubicacion y con bbox_to_anchor se mueve en los ejes
# plt.legend(loc='lower left', bbox_to_anchor=(0.4,0.2))
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
rows=3
cols=3
fig, axes = plt.subplots(3,3, figsize=(12,6))
fig.tight_layout()
for row in range(rows):
    #print(row)#
    for col in range(cols):
        axes[row,col].plot(x, np.sin(x))
        axes[row,col].set_title("sin(x)")
        #print(col)
```
